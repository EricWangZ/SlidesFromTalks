```python
%%latex 
$$ h(X) = \sum_{k=1}^n x_i w_i$$
$$  y(h) = \begin{cases} 1,& \text{if } h\gt 0 \\ 0,& \text{otherwise} \end{cases} $$
```

## Training an MP Neuron
* **All learning happens in the model's weights!**
* Hebb's Rule* - "Neurons that fire together wire together".

```python
%%latex

$$ \Delta w_{i,k} $$
$$ = t_k \times x_i - y_k \times x_i $$
$$ = ( t_k - y_k ) \times x_i $$
$$  $$
$$ w_{i,j}^{t+1} = w_{i,j}^{t} + \eta \Delta w_{i,k}  $$
$$ $$
$$ \text{learning rate: } \eta \in (0,1) $$
```

# Activation Functions

```python
def onoff_activation(h):
    return 1 if h > 0 else 0

def sigmoid_activation(h,B=2):
    return 1./(1+np.exp(-B*h))

def tanh_activation(h):
    return (np.exp(h)-np.exp(-h))/(np.exp(h)+np.exp(-h))

figure(9)
title('Origiinal Activation')
X = np.arange(-1,1.0001,.001)
Y = [onoff_activation(x) for x in X]
plot(X,Y)
xlim([-1,1])
ylim([-.05,1.05])

figure(10)
title('Sigmoid')
X = np.arange(-5,5.0001,.001)
Y = [sigmoid_activation(x) for x in X]
plot(X,Y)
xlim([-5,5])
ylim([-.05,1.05])

figure(11)
title("Hyperbolic Tangent")
Y = [tanh_activation(x) for x in X]
plot(X,Y)
xlim([-5,5])
ylim([-1.05,1.05])
```

### Aside: selecting the number of hidden nodes.

* Another useful theorem exists.  It says that, given two networks with equal perceived performance, the simpler of the two networks will generalize better.  The idea is that any unnecessary complexity is more likely to cause you to overfit than anything.
* in choosing the number of hidden nodes, I trained and tested networks with hidden nodes ranging from 1 to 12 hidden nodes.  less than 5 hidden nodes saw a significant decrease in performance.  6 or more nodes all gave nearly the same (much better than less than 5 node) performance.  5 nodes gave almost the performance of any of the 6+ hidden node networks, so I went with 5 hidden nodes.

## Network performance, on many different train/test splits using 5 hidden nodes:
* Average train classification rate: 91.25%
* Average test set classification: 78.00%

# The experiment - determining variable "importance":
* For 200 times:
* * pull a new train/test set
* * train a new network.
* * test network and record base performance
* * for each input variable v:
* * * set v to 0 in all test samples.
* * * retest network and compare to base performance.  record difference
* * * set v to 1 in all test samples.  retest. record difference
* graph findings.  (the more "important" a variable, the more it's removal will hurt performance.
